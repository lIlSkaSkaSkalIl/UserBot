# üîê Generate Session String (Pyrogram v2)

Cell ini digunakan **hanya sekali saja** untuk menghasilkan `SESSION_STRING`.  
Anda cukup menjalankan cell ini, lalu **salin hasil session string-nya** dan tempelkan ke dalam `config.py` atau environment variable.

‚ö†Ô∏è **Jangan jalankan ulang setiap kali ingin menjalankan bot.**

```python
# @title <center><font color="red">Generate Session String</font></center> { display-mode: "form" }

# Masukkan kredensial Anda
API_ID = None  # @param {type:"integer"}
API_HASH = ""  # @param {type:"string"}

# ‚úÖ Install Pyrogram v2 dan tgcrypto
!pip install -U pyrogram tgcrypto --quiet

# üîê Login Telegram dan export session string (v2 compatible)
import nest_asyncio
import asyncio
from pyrogram import Client

nest_asyncio.apply()  # untuk mencegah error loop saat di Colab

async def main():
    async with Client("gen_session", api_id=API_ID, api_hash=API_HASH) as app:
        session_string = await app.export_session_string()
        print("\n‚úÖ SESSION_STRING berhasil dibuat:\n")
        print(session_string)

# Jalankan
await main()
```
